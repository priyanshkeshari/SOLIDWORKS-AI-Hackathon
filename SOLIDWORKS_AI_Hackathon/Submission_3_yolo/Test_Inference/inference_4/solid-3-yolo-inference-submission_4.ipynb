```python
import glob
import cv2

model = YOLO(BEST_MODEL_PATH)
test_images = glob.glob(os.path.join(DATA_DIR, "test/test", "*.png"))

print(f"Processing {len(test_images)} images with Median Voting TTA...")

submission_rows = []

# Define specific thresholds per class to optimize precision/recall
# Washers are hard to see -> Lower threshold (0.15)
# Bolts are easy -> Higher threshold (0.50) to avoid false positives
THRESHOLDS = {
    'bolt': 0.50,
    'locatingpin': 0.35,
    'nut': 0.35,
    'washer': 0.15 
}

for img_path in test_images:
    img_cv = cv2.imread(img_path)
    
    # Create 3 versions of the image
    img_variants = [
        img_cv,                         # 1. Original
        cv2.flip(img_cv, 1),            # 2. Horizontal Flip
        cv2.flip(img_cv, 0)             # 3. Vertical Flip
    ]
    
    votes = {'bolt': [], 'locatingpin': [], 'nut': [], 'washer': []}
    
    # Run inference on all 3 versions
    for variant in img_variants:
        results = model(variant, verbose=False, iou=0.6)[0]
        
        # Count for this variant
        temp_counts = {'bolt': 0, 'locatingpin': 0, 'nut': 0, 'washer': 0}
        
        if results.boxes:
            for box in results.boxes:
                cls_id = int(box.cls)
                cls_name = model.names[cls_id]
                conf = float(box.conf)
                
                # Use Per-Class Thresholds
                if conf >= THRESHOLDS.get(cls_name, 0.25):
                    temp_counts[cls_name] += 1
        
        # # Logic Filter: Cap at 4 (apply to each vote individually)
        for k in temp_counts:
            # if temp_counts[k] > 4: temp_counts[k] = 4
            votes[k].append(temp_counts[k])
            
    # CALCULATE MEDIAN for each class
    # Example: Votes [4, 4, 3] -> Median 4
    # Example: Votes [4, 5, 5] -> Median 5 (but clipped to 4 earlier) -> 4
    final_counts = {}
    for k in votes:
        final_counts[k] = int(np.median(votes[k]))

    submission_rows.append({
        "image_name": os.path.basename(img_path),
        "bolt": final_counts['bolt'],
        "locatingpin": final_counts['locatingpin'],
        "nut": final_counts['nut'],
        "washer": final_counts['washer']
    })

# SAVE
df_sub = pd.DataFrame(submission_rows)
df_sub = df_sub.reindex(columns=['image_name', 'bolt', 'locatingpin', 'nut', 'washer'], fill_value=0)
df_sub.to_csv("submission_median_tta.csv", index=False)
print("Saved submission_median_tta.csv")
```
